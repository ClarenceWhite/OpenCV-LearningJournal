# Annotating Images
- Draw lines
- Draw circles
- Draw rectangles
- Add text

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)
from IPython.display import Image
```

```python
#Read the image
img1 = cv2.imread("1.png", cv2.IMREAD_COLOR)
img1 = img1[:, :, ::-1] #convert color channel

plt.imshow(img1)
```

# Drawing a line

```python
#make a copy of the image
img1_copy = img1.copy()
cv2.line(img1_copy, (500,500), (2000,500), (0, 255, 255), thickness=10, lineType=cv2.LINE_AA)
plt.imshow(img1_copy)
```

# Drawing a circle

```python
#make another copy of the img
img1_copy2 = img1.copy()
cv2.circle(img1_copy2, (1850, 1190), 100, (255,0,0), thickness=5, lineType=cv2.LINE_AA)
plt.imshow(img1_copy2)
```

# Drawing a rectangle

```python
#make another copy of the img
img1_copy3 = img1.copy()
cv2.rectangle(img1_copy3, (1050,80), (1350, 690), (255,0,0), thickness=5, lineType=cv2.LINE_8)
plt.imshow(img1_copy3)
```

# Adding text

```python
#mak another copy of the image
img1_copy4 = img1.copy()
text = "Musk's spaceX, so cool!"
fontScale = 11
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (255,0,0)
fontThickness = 13

cv2.putText(img1_copy4, text, (500, 1600), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA)
plt.imshow(img1_copy4)
```
