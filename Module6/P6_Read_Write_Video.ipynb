# Writing a video using openCV
While building applications, it becomes important to save a video of your work as well as many applications might require saving a video clip.

## Read video from source

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline

source  = './video.mp4'
cap = cv.VideoCapture(source)
```

```python
if (cap.isOpened() == False):
    print("Error opening video stream or file")
```

## Read and display one frame

```python
ret, frame = cap.read()
```

```python
plt.imshow(frame[..., ::-1])
```

## Display the video from file

```python
from IPython.display import HTML
HTML(
    """
    <video width=1024 controls>
        <source src="video.mp4"  type="video/mp4">
    </video>
    """
)
```

## Write video using OpenCV

```python
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# define the codec and create   VideoWriter object
out_avi = cv.VideoWriter('source.avi', cv.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width, frame_height))

out_mp4 = cv.VideoWriter('source.mp4', cv.VideoWriter_fourcc(*'mp4v'), 10, (frame_width, frame_height))
```

## Read frames and write to file

```python
while(cap.isOpened()):
    ret, frame = cap.read()

    if ret == True:
        out_avi.write(frame)
        out_mp4.write(frame)
    else:
        break
```

```python
cap.release()
out_avi.release()
out_mp4.release()
```
