# Basic image manipulations
- Accessing and manipulation image pixels
- Image resizing
- Cropping
- Flipping

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import Image
%matplotlib inline
```

# Read the image

```python
img1 = cv2.imread("1.png", 0)

#set color map to gray scale for proper rendering
plt.imshow(img1, cmap='gray')
print(img1)
```

# Accessing individual pixels

```python
#print the first pixel of the first black box
print(img1[0, 0])
#print the first white pixel to the right of the first black box
print(img1[0, 6])
```

# Modifying Image Pixels

```python
img1_copy = img1.copy()
for j in range(50):
    for i in range(50):
        img1_copy[j,i] = 200
        img1_copy[j,i] = 200

plt.imshow(img1_copy, cmap='gray')
print(img1_copy)
```

# Cropping

```python
img2 = cv2.imread("2.png", cv2.IMREAD_COLOR)
img2 = img2[:, :, ::-1] #change BGR to RGB
print(img2.shape)

plt.imshow(img2)
```

```python
img2_cropped = img2[0:500, 200:1000]
plt.imshow(img2_cropped)
```

# Resizing

## Method 1: Specifying scaling factor using fx and fy

```python
img2_cropped_resized = cv2.resize(img2_cropped, None, fx=2, fy=2)
plt.imshow(img2_cropped_resized)
#we can see that the size of the cropped image has doubled
```

## Method 2: Specifying exact size of the output image

```python
width = 400
height = 1000
dim = (width, height)

#resize
img2_cropped_resized2 = cv2.resize(img2_cropped, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(img2_cropped_resized2)
```

## Resize while maintaining aspect ratio

```python
width = 100
aspect_ratio = width/img2_cropped.shape[1]
height = int(img2_cropped.shape[0]*aspect_ratio)
dim = (width, height)

#resize
img2_cropped_resized3 = cv2.resize(img2_cropped, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(img2_cropped_resized3)
```

# Flipping

```python
img2_flipped_horz = cv2.flip(img2, 1)
img2_flipped_vert = cv2.flip(img2, 0)
img2_flipped_both = cv2.flip(img2, -1)

#show
plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(img2_flipped_horz); plt.title("Horizontal flip")
plt.subplot(142); plt.imshow(img2_flipped_vert); plt.title("Vertical flip")
plt.subplot(143); plt.imshow(img2_flipped_both); plt.title("Both flip")
plt.subplot(144); plt.imshow(img2); plt.title("Original")
```
